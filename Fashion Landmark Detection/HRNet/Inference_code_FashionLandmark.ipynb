```python
#custome "data/deepfashion2/validation/image" --> 1000 images 

import glob
import os 
import shutil 

src_dir = '/content/HRNet-for-Fashion-Landmark-Estimation.PyTorch/data/deepfashion2/validation/image'
dst_dir = '/content/HRNet-for-Fashion-Landmark-Estimation.PyTorch/data_custom/image'
count = 0
for jpgfile in sorted(glob.glob(os.path.join(src_dir, "*.jpg"))): 
  shutil.copy(jpgfile, dst_dir)
  count += 1
  if count == 1000: 
    break

print('done')
num_img = os.listdir("/content/HRNet-for-Fashion-Landmark-Estimation.PyTorch/data_custom/image")
print("number of copied images: ", len(num_img))
```
