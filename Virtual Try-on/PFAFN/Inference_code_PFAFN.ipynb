```python
img_path = '/content/PF-AFN/PF-AFN_test/results/demo1/PFAFN'
img_per_path = '/content/result/img'
img_cloth_path = '/content/result/clothes'
img_out_path = '/content/result/mask'
```

```python
imgtest = Image.open('/content/PF-AFN/PF-AFN_test/results/demo1/PFAFN/0.jpg')
print(imgtest.size)
data = np.asarray(imgtest)
p = data[:, :192]
print("p: ",p.shape)
c = data[:, 192:384]
print("c: ",c.shape)
o = data[:,384:]
print("o: ",o.shape)
print(data.shape)
plt.imshow(p)
plt.imshow(c)
plt.imshow(o)
```

```python
for i in sorted(glob.glob(os.path.join(img_path, "*.jpg"))):
  img = Image.open(i)
  d = np.asarray(img)
  name = os.path.basename(i).split('.')[0]

  person = d[:, :192]
  per_img = Image.fromarray(person)
  per_save = per_img.save(os.path.join(img_per_path,name+".jpg"))

  cloth = d[:,192:384]
  cl_img = Image.fromarray(cloth)
  cl_save = cl_img.save(os.path.join(img_cloth_path,name+".jpg"))

  out = d[:, 384:]
  out_img = Image.fromarray(out)
  out_save = out_img.save(os.path.join(img_out_path,name+".jpg"))


print(len(os.listdir(img_cloth_path)))
print(len(os.listdir(img_out_path)))
print(len(os.listdir(img_per_path)))

```
